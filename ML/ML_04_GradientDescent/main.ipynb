```python
import matplotlib.pyplot as plt

# function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

# parameters
x = 2               # starting point
alpha = 0.1         # learning rate
epochs = 25         # number of iterations

x_values = [x]
y_values = [f(x)]

# gradient descent loop
for i in range(epochs):
    grad = df(x)
    x = x - alpha * grad
    x_values.append(x)
    y_values.append(f(x))
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

# plot the results
plt.plot(x_values, y_values, marker='o')
plt.title("Gradient Descent: Finding Local Minima of y = (x + 3)^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

print(f"\nLocal minimum occurs at x = {x:.4f}")
```
