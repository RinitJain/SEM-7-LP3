```python
import pandas as pd
import numpy as np
import seaborn as sns
ds = pd.read_csv("sales_data_sample.csv", encoding='Latin-1')
ds.head()
```

```python
ds.dtypes
```

```python
ds.isna().sum()
```

```python
# Convert ORDERDATE to numeric
ds['ORDERDATE'] = pd.to_datetime(ds['ORDERDATE']).astype('int64')

# Encode STATUS column using mapping
status_map = {
    'Shipped': 1,
    'Resolved': 2,
    'Cancelled': 3,
    'On Hold': 4,
    'Disputed': 5,
    'In Process': 6
}
ds['STATUS'] = ds['STATUS'].map(status_map)

# Drop unnecessary columns
cols_to_drop = [
    'PRODUCTCODE',
    'ADDRESSLINE1', 'ADDRESSLINE2', 'POSTALCODE', 'TERRITORY',
    'CITY', 'STATE', 'COUNTRY',
    'CUSTOMERNAME', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME',
    'DEALSIZE', 'PHONE'
]
ds.drop(columns=cols_to_drop, inplace=True)

# One-hot encode PRODUCTLINE
ds = pd.get_dummies(ds, columns=['PRODUCTLINE'])
```

```python
ds.describe()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dataset = ds.values

WCSS = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(dataset)
    WCSS.append(kmeans.inertia_)

plt.plot(range(1,20), WCSS)
```

```python
#Choosing cluster size n=3

kmeans = KMeans(n_clusters=3)
kmeans.fit(dataset)
labels = kmeans.predict(dataset)
centroids = kmeans.cluster_centers_
```
